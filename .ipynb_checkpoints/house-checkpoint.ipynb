```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline 
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
plt.scatter(df.area,df.price,color='red')
plt.plot(df.area,reg.predict(df[['area']]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv('canada_per_capita_income.csv')
d
```

```python
income = d['per capita income (US$)']
plt.scatter(d.year,income)
```

```python
regi = linear_model.LinearRegression()
regi.fit(d[['year']],income)
```

```python
regi.predict([[2025]])
```

```python
plt.plot(d.year,regi.predict(d[['year']]))
plt.scatter(d.year,income,color='red',marker='+')
```
